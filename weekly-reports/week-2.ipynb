There are two main protocol distinctions: *Bluetooth Classic* and *Bluetooth Low Energy (BLE)*. __Any version below version 4 is considered bluetooth classic__.

### Bluetooth Low Energy (BLE)
- Low bandwidth applications like sensor data
- Optimized for low power
- Operates over 40 RF channels
- Discovery occurs over 3 channels, quicker connection
- Laungth with Bluetooth 4.0 in early 2010
- Bluetooth v4.2 provided security upgrades, IP connectivity, and increased speed in 2014
- Bluetooth v5.0 provided 2x the speed and 4x the range of Bluetooth 4

### Bluetooth Classic
- High bandwidth applications like streaming audio
- Not optimized for low power
- Operates over 79 RF channels
- Discovery occurs over 32 channels
- Intrused first in May 1998 and the first specification was published in 1999.
- Bluetooth 3.0 introduced high-speed transfers and allowed devices to transfer files over Wi-Fi while simulataneously communication over Bluetooth in 2009.

### Bluetooth Stack Splits

![BLE stack](./ble-stack.jpeg)

There are three main blocks of the architecture:
- Application
  - Uses the host software to build use cases
- Host
  - Software that manages how two or more devices communicate
  - Firmware we program that controls the device
- Controller
  - Physical device that can transmit and receive radio signals

Each block is made of layers.

### Different solutions are available
Single-Chip Solution
- The application, host, and controller blocks are all on the same chip

Two-Chip Solution
- The application block and the host block are in the same chip while the controller block is on a seperate chip
- The host block and controller blocks are on the same chip while the application block is on a seperate chip __*(this is the most popular configuration)*__

Three-Chip Solution
- Application, controller, and host blocks are all on separate chips.

For our projects we will be using the two-chip solution.
